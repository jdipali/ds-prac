```python
import math
print('Using 3 inputs, 3 weights, 1 output')
x1 = [0.3, 0.5, 0.8]
w1 = [0.1, 0.1, 0.1]
t = 1
a = 0.1
diff = 1
yin = 0
while(diff > 0.4):
    for i in range(0, 3):
        yin = yin + (x1[i] * w1[i])
    yin = yin + 0.25
    yin = round(yin, 3)
    print('Yin: ', yin)
    print('Target: ', t)
    diff = t - yin
    diff = round(diff, 3)
    diff = math.fabs(diff)
print('Error: ', diff)
new1 = []
for i in range(0, 3):
    wnew1 = w1[i] + a * diff * x1[i]
    wnew1 = round(wnew1, 2)
    new1.append(wnew1)
w1 = new1
print('w1new=', w1)
```
